```python
import glob2
import os
import random
all_files = os.listdir('/content/drive/MyDrive/Deep_Learning/Yolov6/yolo_plate_ocr_dataset/img/train')
random_no_repeat = random.sample(range(len(all_files)),int(0.2*len(all_files)))
all_files[1]
```

```python
import shutil

path_jpg = '/content/drive/MyDrive/Deep_Learning/Yolov6/yolo_plate_ocr_dataset/img/train/'
path_txt = '/content/drive/MyDrive/Deep_Learning/Yolov6/yolo_plate_ocr_dataset/label/train/'
move_jpg = '/content/drive/MyDrive/Deep_Learning/Yolov6/yolo_plate_ocr_dataset/img/val'
move_txt = '/content/drive/MyDrive/Deep_Learning/Yolov6/yolo_plate_ocr_dataset/label/val'

for i in range(len(random_no_repeat)):
  path1 = path_txt + all_files[random_no_repeat[i]][:-3] + 'txt'
  # print(path1)
  path2 = path_jpg + all_files[random_no_repeat[i]]
  if os.path.exists(path1):
    shutil.move(path1,move_txt)
    shutil.move(path2,move_jpg)
```
