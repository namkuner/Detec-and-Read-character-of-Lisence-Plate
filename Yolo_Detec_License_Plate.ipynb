```python
import os
os.chdir('/content/drive/MyDrive/Deep Learning/Yolo_Practice/darknet')
print(os.getcwd())
import numpy as np
import glob2
files_jpg = glob2.glob(os.path.join('/content/drive/MyDrive/Deep Learning/Yolo_Practice/darknet/data/yolo_plate_dataset/',"*.jpg"))
len(files_jpg)
```

```python
import random

random_no_repeat = random.sample(range(len(files_jpg)),int(0.2*len(files_jpg)))
len(random_no_repeat)
```

```python
random_no_repeat
```

```python
import os
train_file = open('train.txt','w')
valid_file = open('test.txt','w')

for ordinal_number in range(len(files_jpg)):
  if ordinal_number  not in random_no_repeat and os.path.exists(files_jpg[ordinal_number][:-3]+'txt'):
    train_file.write(files_jpg[ordinal_number] + '\n')
  elif ordinal_number in random_no_repeat and os.path.exists(files_jpg[ordinal_number][:-3]+'txt'):
    valid_file.write(files_jpg[ordinal_number] + '\n')
train_file.close()
valid_file.close()
```
